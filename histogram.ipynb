```python
import chart_studio.plotly as py
import plotly.graph_objs as go
import plotly.offline as offline
import numpy as np
offline.init_notebook_mode(connected=True)
```

```python
random_numbers = np.random.randn(200)
```

```python
hist = go.Histogram(x = random_numbers)
```

```python
data = [hist]
offline.iplot(data)
```

```python
# plot using probability density 
hist = go.Histogram(x = random_numbers, histnorm='probability')
```

```python
data = [hist]
offline.iplot(data)
```

```python
# cumulative histogram incllude data points from all previous ranges in each bin 
# monotonically increasing
```

```python
hist = go.Histogram(x=random_numbers,
                    cumulative=dict(enabled=True)
                   )
```

```python
data = [hist]
offline.iplot(data)
```

```python
more_random_numbers = 1 + np.random.randn(100)
```

```python
trace0 = go.Histogram(
    x=random_numbers,
    opacity=0.9
)
```

```python
trace1 = go.Histogram(
    x=more_random_numbers,
    opacity=0.5
)
```

```python
data = [trace0,trace1]
layout = go.Layout(barmode='overlay')
```

```python
fig = go.Figure(
    data=data,
    layout=layout
)
offline.iplot(fig)
```

```python
trace0 = go.Histogram(
    x = random_numbers,
    name = 'Series 1',
    xbins = dict(start=-4,end=4,size=0.5),
    marker = dict(color='#D67D93'),
    opacity=0.75
)
```

```python
trace1 = go.Histogram(
    x = more_random_numbers,
    name = 'Series 2',
    xbins = dict(start=0,end=4,size=0.5),
    marker = dict(color='#7DAED6'),
    opacity=0.75
)
```

```python
data = [trace0,trace1]
layout = go.Layout(
    title='Sytled histograms',
    xaxis=dict(title='Value'),
    yaxis=dict(title='Count'),
    bargap=0.2,
    bargroupgap=0.1
)
```

```python
fig = go.Figure(data=data,layout=layout)
offline.iplot(fig)
```
