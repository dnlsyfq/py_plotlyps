```python
import plotly.graph_objs as go
import plotly.offline as offline
import numpy as np
import pandas as pd 
offline.init_notebook_mode(connected=True)
```

```python
n = 200

random_x = np.random.randn(n)
random_y = np.random.randn(n)
```

```python
trace = go.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers'
)

fig = go.Figure(data=[trace])
fig
```

```python
n = 50
random_x = np.linspace(0,1,n)
random_y0 = np.random.randn(n) + 5
random_y1 = np.random.randn(n)
random_y2 = np.random.randn(n) - 5
```

```python
trace0 = go.Scatter(
    x = random_x,
    y = random_y0,
    mode = 'markers'
)

trace1 = go.Scatter(
    x = random_x,
    y = random_y1,
    mode = 'lines'
)

trace2 = go.Scatter(
    x = random_x,
    y = random_y2,
    mode = 'lines+markers'
)
```

```python
fig = go.Figure(data=[trace0,trace1,trace2])
fig
```

```python
honey_data = pd.read_csv('datasets/honeyproduction.csv')
```

```python
honey_data.sample(2)
```

```python
trace = go.Scatter(
    x = honey_data['state'],
    y = honey_data['year'],
    mode = 'markers',
    marker = dict(
        size=14,
        color=honey_data['totalprod'],
        colorscale='Rainbow',
        showscale=True,
        opacity=0.5
    ),
    text = honey_data['totalprod']
)
```

```python
data = [trace]
```

```python
layout = go.Layout(
    title='Honey production the United States',
    hovermode='closest',
    xaxis=dict(title='States'),
    yaxis=dict(title='Honey Production')
)
```

```python
fig = go.Figure(data,layout)
fig
```
