```python
import plotly.graph_objs as go
import plotly.offline as offline
import numpy as np
import pandas as pd 
offline.init_notebook_mode(connected=True)

from datetime import datetime
pd.core.common.is_list_like = pd.api.types.is_list_like
import pandas_datareader.data as web
```

```python
# pip install pandas_datareader
```

```python
amzn = web.DataReader('AMZN','yahoo',
    datetime(2017,1,1),
    datetime(2018,6,1)
).reset_index()
```

```python
amzn.head()
```

```python
data = [go.Scatter(
    x=amzn.Date,
    y=amzn.High
)]

offline.iplot(data)
```

```python
trace_high = go.Scatter(
    x=amzn.Date,
    y=amzn.High,
    name='Amazon High',
    line=dict(color='#6699FF')
)

trace_low = go.Scatter(
    x=amzn.Date,
    y=amzn.Low,
    name='Amazon Low',
    line=dict(color='#FF6633')
)
```

```python
data = [trace_high,trace_low]

layout = dict(title='Amazon Stock Price Data')
```

```python
go.Figure(data,layout)
```

```python
layout = dict(
    title='Amazon Stock Price Data',
    xaxis=dict(
        rangeslider=dict(),
        type='date'
    )
)
```

```python
go.Figure(data,layout)
```

```python
layout = dict(
    title='Amazon Stock Price Data',
    xaxis=dict(
        rangeslider=dict(),
        rangeselector=dict(buttons=list(
            [dict(count = 1,
                  label = '1m',
                  step = 'month',
                  stepmode = 'todate'
                 ),
             dict(count = 6,
                  label = '6m',
                  step = 'month',
                  stepmode = 'backward'
                 ),
             dict(step = 'all')
            ]
        )),
        type='date'
    )
)
```

```python
go.Figure(data,layout)
```
